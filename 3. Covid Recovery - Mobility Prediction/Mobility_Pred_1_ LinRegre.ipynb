```python
df = pd.read_csv(r"C:\Users\aravindh.saravanan\Desktop\dashboard refresh\20july_input.csv")
df_test = pd.read_csv(r"C:\Users\aravindh.saravanan\Desktop\dashboard refresh\20thjuly_forecasted.csv")
df_test['date'] = pd.to_datetime(df_test['date'])
df['date'] = pd.to_datetime(df['date'])
output=pd.DataFrame()
Error_report=pd.DataFrame(columns=['rmsd','r2_value','intercept','mape'])
for state in df.state.unique():
    print("\n Running for State: "+state)
    data = df[df['state']==state]
    data = data.drop("state",1)
    data = data.drop("date",1)

    x=data.iloc[:,0:6]
    x = x.drop(["workplaces_percent_change_from_baseline"],1,errors='ignore')
    x=x.fillna(method = 'ffill')
    x=x.fillna(method = 'bfill')
    coeff_df=pd.DataFrame(columns=x.columns)
    y=data["gross_ma"]
    model = linear_model.LinearRegression()
    model.fit(x, y)
    coeff_df.loc[state]=model.coef_


    data_test = df_test[df_test['state']==state]
    data_test = data_test.drop("state",1)
    data_test = data_test.set_index("date",drop=True)

    x_test=data_test.iloc[:,0:6]
    x_test = x_test.drop(["workplaces_percent_change_from_baseline"],1,errors='ignore')
    y_test=data_test["gross_ma"]
    x_test=x_test.fillna(method = 'ffill')
    x_test=x_test.fillna(method = 'bfill')
    y_pred = model.predict(x_test)

    out = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
    out['state']=state
    output=output.append(out)
    # Root Mean Squared Deviation
    rmsd = np.sqrt(mean_squared_error(y_test, y_pred))
    r2_value = r2_score(y_test, y_pred)
    intercept=model.intercept_
    mape=( abs((y_test - y_pred) / y_test).mean()) * 100
    value=[rmsd,r2_value,intercept,mape]
    Error_report.loc[state]=value

output.to_csv(r"C:\Users\aravindh.saravanan\Desktop\dashboard refresh\20july_output.csv")
coeff_df.to_csv(r"C:\Users\aravindh.saravanan\Desktop\dashboard refresh\20july_coeff.csv")
Error_report.to_csv(r"C:\Users\aravindh.saravanan\Desktop\dashboard refresh\20july_error.csv")
```
